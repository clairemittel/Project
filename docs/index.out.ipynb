# Introduction/Background/Methods

The topic we have chosen for our project is to see how temperature affects productivity in the Great Lakes. We are going to show the relationship by seeing how the temperature that is recorded in the lakes, Lake Erie and Lake Huron, increase or decrease the amount of Chlorophyll A. Chlorophyll A is able to show that there is high productivity in environments because it is an indicator of how much primary productivity has occurred (Baker & Oxborough ([2004](#ref-baker2004chlorophyll))). Primary productivity is the amount of photosynthesis that is occurring within an ecosystem. A high productivity would be indicated by a high amount of Chlorophyll A because it is what makes plants green and when they are most green, they are able to absorb the most light (Björn et al. ([2009](#ref-bjorn2009viewpoint))). This is an important topic to be discussing because it has been shown that anthropogenic climate change has altered how much primary productivity there is in lakes. There was a study, O’Beirne et al. ([2017](#ref-o2017anthropogenic)), that showed that primary productivity has decreased very slowly in the last 9,000 years. Primary productivity is important to lake environments because the photosynthesis that occurs produces nutrients for consumers to thrive off of. When primary productivity goes down, the overall health of the lake goes down. The motivation behind seeing if there is a correlation between the two is that the lakes are two very different ecosystems. Another reason behind choosing this data is that Lake Erie is the most polluted of the Great Lakes. This could have an effect on how much productivity there is in the lake because there could be less light able to penetrate the water, and therefore a lower temperature for a possible lower productivity rate.

The datasets that we will be using are from the NOAA Great Lakes Environmental Research Laboratory, US Department of Commerce ([n.d.-b](#ref-lakeerie)) and US Department of Commerce ([n.d.-a](#ref-lakehuron)). The datasets are from the same laboratory that is studying the Great Lakes and taking measurements to keep track of the water quality in each of the lakes. In Lake Erie and Lake Huron, they are keeping track of Microcystis, which is a “toxin producing algae” (US Department of Commerce ([n.d.-b](#ref-lakeerie))). They are keeping track of the amount of Microcytis to see if there is a correlation between the amount of it in the water and the toxicity level in the lake. They collect other weekly environmental parameters including “nitrate (N), chlorophyll, phycocyanin, dissolved oxygen (DO), turbidity, and specific conductivity” (US Department of Commerce ([n.d.-b](#ref-lakeerie))). With this data they are collecting for tracking Microcystis, they are looking to figure out what causes the algal blooms so that they are able to prepare for the harm they will cause. One method that we will be using in this project to determine the relationship between temperature and Chlorophyll A levels is a Chi squared test. This test will help us determine if the relationship between temperature and Chlorophyll A levels are dependent on each other or not. This will help us find a correlation because the expected results would be that there is no correlation and if it is different from the expected, it proves that there is a correlation. Another method that we will be using in this project to determine the relationship between temperature and Chlorophyll A levels is a correlation test. This will help to see if there is a linear correlation between temperature and Chlorophyll A levels by seeing if they have a strong relationship. This will help see what the relationship is because if there is a strong correlation, it proves that the relationship is either positive or negative. It is most likely a positive relationship because warmer temperatures tend to increase the amount of photosynthesis that occurs (@dunstan2018global). That means that an increase of photosynthesis means that there is an increase in Chlorophyll A. Therefore, there would be a positive relationship between temperature and Chlorophyll A levels.

# Methods

A Chi squared test will be administered to test if there is a relationship between temperature and Chlorophyll A levels for each dataset. We will also be using a correlation test to find if there is a linear correlation between temperature and Chlorophyll A for each dataset. The two datasets themselves will be tested to determine if there are significant differences between measures of Chlorophyll a between the lakes. If data are normal and variances equal we will use a two-sample t-test (independent t-test). If data are not normal we will use a non-parametric test, like the Wilcoxon Rank-Sum Test. There will also be a test to find if there is a significant difference in the way temperature and Chlorophyll a interact in the different lakes. For this test we will use a linear model with an interaction between lake and temperature.

Some supplemental data we may need is information on why the 2 lakes may differ in Chlorophyll a levels or their relationship with Chlorophyll a and temperature. We might need data on pollution levels to each lake. The data seems like it will be easy to work with but Dataset 1 and Dataset 2 have a very different amount of rows. To make certain tests work we may need to condense the data from Dataset 1. It also might be helpful to only include data from 5/22-10/18 because this is the time range of data collected in Dataset 1. Some potential challenges we see in the methods is understanding when to use certain statistical methods/making sure they are the best statistical method for the data and research question.

These selected methods help us get to our final goal because in various ways we will test the correlation between temperature and CHLa (productivity). This information will be valuable because as temperatures increase, productivity may increase and cause problems for lake ecosystems. We will also compare Lake Erie to Lake Huron because pollution to lake Erie may affect productivity levels or the way temperature impacts productivity.

```
lakeerie2 <- lakeerie %>%
  rename(
    surface_temp = Temp_C,
    chlorophyll_a = `Extracted_CHLa_ugL-1`)
lakehuron2 <- lakehuron %>%
  rename(
    surface_temp = Temp_C,
    chlorophyll_a = `Extracted_CHLa_ugL-1`)
```

## Testing

```
ggplot(lakeerie2, aes(x = surface_temp, y = chlorophyll_a)) +
  geom_point(alpha = 0.4) +
  geom_smooth(method = "lm", color = "blue") +
  labs(title = "Temperature vs Chlorophyll a in Lake Eerie",
       x = "Surface Temperature (°C)",
       y = "Chlorophyll a (µg/L)")
```

```
ggplot(lakehuron2, aes(x = surface_temp, y = chlorophyll_a)) +
  geom_point(alpha = 0.4) +
  geom_smooth(method = "lm", color = "lightblue") +
  labs(title = "Temperature vs Chlorophyll a in Lake Huron",
       x = "Surface Temperature (°C)",
       y = "Chlorophyll a (µg/L)")
```

```
library(tidyverse)

lakeerie_clean <- lakeerie2 %>%
  drop_na(surface_temp, chlorophyll_a) %>%
  mutate(
    surface_temp = as.numeric(surface_temp),
    chlorophyll_a = as.numeric(chlorophyll_a))
```

### Lake Eerie Tests

#### Chi Squared Test

```
lakeerie_binned <- lakeerie_clean %>%
  mutate(
    temp_cat = ntile(surface_temp, 3),
    chl_cat = ntile(chlorophyll_a, 3)
  ) %>%
  mutate(
    temp_cat = factor(temp_cat, labels = c("Low", "Medium", "High")),
    chl_cat = factor(chl_cat, labels = c("Low", "Medium", "High"))
  )

table_data <- table(lakeerie_binned$temp_cat, lakeerie_binned$chl_cat)

chisq.test(table_data)
```

#### Pearson’s Correlation Test

```
cor.test(lakeerie_clean$surface_temp, lakeerie_clean$chlorophyll_a, method = "pearson")
```

#### T-Test

```
chl_high <- lakeerie_binned %>% filter(temp_cat == "High") %>% pull(chlorophyll_a)
chl_low  <- lakeerie_binned %>% filter(temp_cat == "Low") %>% pull(chlorophyll_a)

t.test(chl_high, chl_low, var.equal = FALSE)
```

#### Wilcox Test

```
wilcox.test(chl_high, chl_low)
```

### Lake Huron Testing

#### Chi Squared Test

```
lakehuron_binned <- lakehuron_clean %>%
  mutate(
    temp_cat = ntile(surface_temp, 3),
    chl_cat = ntile(chlorophyll_a, 3)
  ) %>%
  mutate(
    temp_cat = factor(temp_cat, labels = c("Low", "Medium", "High")),
    chl_cat = factor(chl_cat, labels = c("Low", "Medium", "High"))
  )

table_data <- table(lakehuron_binned$temp_cat, lakehuron_binned$chl_cat)

chisq.test(table_data)
```

#### Pearson’s Correlation Test

```
cor.test(lakehuron_binned$surface_temp, lakehuron_binned$chlorophyll_a, method = "pearson")
```

#### T-Test

```
chl_high2 <- lakehuron_binned %>% filter(temp_cat == "High") %>% pull(chlorophyll_a)
chl_low2  <- lakehuron_binned %>% filter(temp_cat == "Low") %>% pull(chlorophyll_a)

t.test(chl_high2, chl_low2, var.equal = FALSE)
```

#### Wilcox Test

```
wilcox.test(chl_high2, chl_low2)
```

# Results

The Chi squared test that was done on the Lake Eerie data (dataset 1), the p value came back as 1.509e-07. This means that there is a significant relationship between temperature and chlorophyll A levels in Lake Eerie. The Pearson correlation test that was performed on the data came back with a p value of 3.013e-10. This also shows a highly significant relationship between temperature and chlorophyll A levels. The T-test done also came back with a very low p value of 3.669e-16. This test also shows a high significance in the relationship of the variables. The final test performed on the data, the Wilcox test, had a p value of 1.092e-15. This very low p value shows that there is a high significance in the relationship between the variables.

The Chi squared test that was performed on the Lake Huron data (dataset 2) the p value came back as 0.03894. This means that the data is significant and there is a relation in this data of temperature to chlorophyll A levels. The Pearson correlation test that was done on the Lake Huron data had a p value of 0.1028. This also shows that there is not a significant linear relationship between temperature and chlorophyll A levels in Lake Huron. The T-test that was done on the data came back with a p value of 0.01676. This shows that there is a significant relationship between the two variables in Lake Huron in this test. The Wilcox test that was performed on the data came back with a p value of 0.0172. This p value means that there is a slightly significant relationship in this data with temperature and chlorophyll A levels.

# Discussion

The Lake Eerie dataset showed much stronger and consistent correlation between temperature and Chlorophyll A levels. This indicates there is a high significance in the relationship between variables and chlorophyll A levels are highly dependent on temperature. For the Lake Huron Dataset, the p-value for the chi-squared test came back as significant while the value for the Pearson Correlation test indicates there was not a significant relationship between temperature and chlorophyll A. Since a chi-test measures association between categorical variables, continuous data is sorted into categories. This suggests there is a non-random association between the categories (high temperature and high chlorophyll A concentration for example). The Pearson correlation tests the strength and direction of the linear relationship between two continuous variables. This indicates that the relationship may not be linear, but could be curved, thresholded or influenced by other variables the Pearson Correlation test cannot detect.

These findings imply that temperature and chlorophyll A have a positive identifiable correlation. Since chlorophyll A is an indicator of high levels of algae productivity, this suggests that when temperatures increase, algal blooms also increase. Large algal blooms can be detrimental to lake ecosystem health. It is helpful to study this relationship because as temperatures increase with climate change, this might increase the level of harmful algal blooms in lake ecosystems. This could influence management decisions and the regulation of pollutants or nutrient input to the great lakes. 

This analysis is very limited because the data used in these statistical tests were collected only during one year. This means that these statistical tests did not account for variation of different factors between years such as climate, nutrient input, and other factors. There was also a fairly small sample size which could have skewed the results. The next steps proposed are to conduct a much longer term study that spans multiple years to account for the variation between years and measure how increasing temperatures are changing chlorophyll A concentration. Using different study sites on different lakes or different parts of the lakes could also give a more comprehensive picture of this relationship.
